```python
import re
import pandas as pd
```

```python
# 1. Kamus Formalisasi (update berkala)
# 'Normalization Lexicon'
slang_to_formal_dict = {
    "gk": "tidak", "gak": "tidak", "ga": "tidak", "nggak": "tidak", "kaga": "tidak",
    "yg": "yang", "utk": "untuk", "dgn": "dengan", "sdh": "sudah", "udah": "sudah",
    "blm": "belum", "bgt": "banget", "tapi": "tetapi", "tp": "tetapi", "klo": "kalau",
    "kalo": "kalau", "lu": "kamu", "gue": "saya", "gw": "saya", "ane": "saya",
    "cuan": "untung", "loss": "rugi", "serok": "beli", "nyangkut": "terjebak",
    "lot": "satuan saham", "emiten": "perusahaan", "skrg": "sekarang", "krn": "karena",
    "aj": "saja", "aja": "saja", "fixed": "pasti", "jd": "jadi"
}
```

```python
def formalize_and_clean(text):
    # a. Lowercase
    text = str(text).lower()

    # b. Hapus URL, Mention, Hashtag
    text = re.sub(r'http\s+|www\.\S+|@[A-Za-z0-9_]+|#[A-Za-z0-9_]+', '', text)

    # c. Tangani Elongated Words (Huruf berulang)
    # Contoh: "banyaaaakkk" -> "banyakk" (kita sisakan 2 agar kata 'saat' tidak rusak)
    text = re.sub(r'(.)\1+', r'\1\1', text)

    # d. Hapus karakter non-alfabet (Simbol & Angka)
    text = re.sub(r'[^a-z\s]', ' ', text)

    # e. Formalisasi Slang
    words = text.split()
    formalized_words = [slang_to_formal_dict.get(w, w) for w in words]
    text = " ".join(formalized_words)

    # f. Hapus whitespace berlebih
    text = ' '.join(text.split())

    return text
```

```python
try:
    df = pd.read_csv('stock_market.csv')
    print("Memulai proses formalisasi teks...")
    df['Sentence_Formal'] = df['Sentence'].apply(formalize_and_clean)
    print("Preprocessing Selesai!")
    # Menampilkan hasil perbandingan
    display(df[['Sentence', 'Sentence_Formal']].head(10))
except FileNotFoundError:
    print("Error: File 'stock_market.csv' tidak ditemukan. Pastikan sudah di-upload ke Colab.")
```
